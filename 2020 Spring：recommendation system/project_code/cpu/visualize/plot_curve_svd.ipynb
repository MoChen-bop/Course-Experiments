```python
def plot_p_r_curves(dataframes, title=None, save_name='None', c='FF00FF'):
    plt.rcParams['figure.dpi'] = 300
    plt.rcParams['savefig.dpi'] = 300
    plt.rcParams['figure.figsize'] = (5, 5)
    dim = 10
    for dataframe in dataframes:
        plt.plot(dataframe['Precises'], dataframe['Recalls'], label='dim = ' + str(dim))
        dim += 10
    plt.title(title)
    plt.xlabel('Precise', fontsize=12)
    plt.ylabel('Recall', fontsize=12)
    plt.yticks(fontproperties = 'Times New Roman', size = 12)
    plt.xticks(fontproperties = 'Times New Roman', size = 12)
    plt.legend(fontsize=10)
    #plt.savefig('./' + save_name + '.jpg', bbox_inches = 'tight')
    #plt.savefig('./' + save_name + '.eps', dpi=300, format='eps', bbox_inches = 'tight')
```

```python
def get_train_loss(file_path):
    steps = []
    losses = []
    for i in range(1, 8):
        file_name = os.path.join(file_path, str(i) + '_train_result.log')
        with open(file_name) as f:
            for line in f.readlines():
                epoch, step, loss, _ = line.split(',')
                steps.append((int(epoch) - 1) * 10000 + int(step) )
                losses.append(float(loss))
    return steps, losses
```

```python
def get_test_auc(file_path):
    steps = []
    aucs = []
    for i in range(1, 8):
        file_name = os.path.join(file_path, str(i) + '_infer_result.log')
        with open(file_name) as f:
            for line in f.readlines():
                auc = line.split(':')[-1][:-2]
                steps.append(i)
                aucs.append(float(auc))
    return steps, aucs
```

```python
deepfm_steps, deepfm_losses = get_train_loss('../results/DeepFM_1_256/DeepFM_1_256')
dnn_steps, dnn_losses = get_train_loss('../results/DNN_1_256/DNN_1_256')
dnnplus_steps, dnnplus_losses = get_train_loss('../results/DNNPlus_1_256/DNNPlus_1_256')
```
